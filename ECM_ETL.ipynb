```python
ecm_files = os.listdir(dir_path)
```

```python
# shell-style pattern in regular expression form used for re.match()
pattern = r'.*\.xlsm\Z'
```

```python
for ecm_filename in ecm_files:
    # test whether filename string matches pattern string (*.xlsm)
    if fnmatch.fnmatch(ecm_filename, '*.xlsm'):
        print str(ecm_filename[:8])
        x = fnmatch.translate('*.xlsm')
        print x
```

```python
## Transform functions
def ecm_clean(df):
    df.dropna(1,'all', inplace=True)
    df.dropna(0,'any', inplace=True)
```

```python
ecm = None
for ecm_filename in ecm_files:
    print ecm_filename
    if not re.match(pattern, ecm_filename):
        continue
    # Get NCP number
    ncp = str(ecm_filename[:8])
    # Get ecm data
    data = pd.read_excel(ecm_filename, 'Supplemental Tables', header=11, parse_cols="C:Q", skip_footer=80)
    # Clean data
    ecm_clean(data)
    # Add NCP number
    data['NCP'] = ncp
    # Get project details
    general = pd.read_excel(ecm_filename, 'General Info')
    ta = pd.read_excel(ecm_filename, 'Contacts')
    # Add project name, city, building type, TA, building area
    data['ProjName'] = str(general.iloc[3,7])
    data['City'] = str(general.iloc[6,7])
    data['BuildType'] = str(general.iloc[14,7])
    data['BuildArea'] = int(general.iloc[11,7])
    data['TA'] = str(ta.iloc[36,1])
    # Save
    if ecm is None:
        ecm = data
    else:
        ecm = ecm.append(data)
```

```python
## Add calculated fields, savings and $ per sf
ecm.drop('Hide Column Q', axis=1, inplace=True)
ecm = ecm.fillna(0)
ecm['Yearly kWh Savings_100sf']=ecm['Yearly kWh Savings']/ecm['BuildArea']*100
ecm['SumPeak_kW_100sf'] = ecm['Customer Summer Peak kW \nSavings']/ecm['BuildArea']*100
ecm['FossilFuel_MMBtu_100sf'] = ecm['Fossil Fuel\nSavings\n(MMBtu)']/ecm['BuildArea']*100
ecm['TotalSave_MMBTU_100sf'] = ecm['Total MMBtu Savings']/ecm['BuildArea']*100
ecm['YearSave_kWh_100sf'] = ecm['Yearly kWh Savings']/ecm['BuildArea']*100
ecm['Inctv_$_100sf'] = ecm['NYSERDA\nIncentive\n($)']/ecm['BuildArea']*100
```
